# Comaparison of r2 scores of various Regression Algorithms 

```python
from sklearn.datasets import make_regression
from sklearn.linear_model import ElasticNet, LinearRegression,Ridge,Lasso
from sklearn.metrics import r2_score
import matplotlib.pyplot as plt
```

```python
X,y = make_regression(n_samples=1000,n_informative=1,n_features=5,n_targets=1,noise=20,random_state=13)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y, test_size = 0.2,random_state = 2)
```

# Elastic Net


```python

EN = ElasticNet(alpha= 1,l1_ratio=0.5)
EN.fit(X_train,y_train)
EN_pred = EN.predict(X_test)
r2_score(y_test,EN_pred)
```

# Linear Regression

```python
LR = LinearRegression()
LR.fit(X_train,y_train)
LR_pred= LR.predict(X_test)
r2_score(y_test,LR_pred)
```

# Ridge Regression

```python
RR = Ridge(alpha = 1)
RR.fit(X_train,y_train)
RR_pred= RR.predict(X_test)
r2_score(y_test,RR_pred)
```

# Lasso Regression

```python
LaR= Lasso(alpha=1)
LaR.fit(X_train,y_train)
LaR_pred= LaR.predict(X_test)
r2_score(y_test,LaR_pred)
```

```python
import matplotlib.pyplot as plt

# R² scores from your results
models = ["Elastic Net", "Linear Regression", "Ridge Regression", "Lasso Regression"]
scores = [0.7806526404884666, 0.8839731733359216, 0.8840094454288286, 0.8839244869247692]

plt.figure(figsize=(8, 5))
bars = plt.bar(models, scores, color=['#ff9999','#66b3ff','#99ff99','#ffcc99'])

plt.ylim(0.75, 0.90)
plt.ylabel("R² Score")
plt.title("Comparison of R² Scores of Various Regression Algorithms")

# Annotate bars with score values
for bar, score in zip(bars, scores):
    plt.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 0.002, f"{score:.3f}", 
             ha='center', va='bottom', fontsize=10)

plt.show()
```
