```python
from sklearn.datasets import make_classification
import matplotlib.pyplot as plt
import numpy as np
```

```python
X,y = make_classification(n_samples=100, n_features=2,  n_informative=1, n_redundant=0, n_repeated=0, n_classes=2, n_clusters_per_class=1,random_state=41, class_sep=10, hypercube= False)
```

```python
plt.scatter(X[:,0],X[:,1],c=y, s= 100)
```

```python
X.shape
```

```python
def perceptron(X,y):
    X=np.insert(X,0,1,axis = 1)
    weights = np.ones(X.shape[1])
    lr = 0.1

    for i in range (1000):
        j = np.random.randint(0,100)
        y_cap = step(np.dot(X[j],weights))
        weights = weights + lr*(y[j]-y_cap)*X[j]

    return weights[0],weights[1:]
```

```python
def step(z):
    return 0 if z<0 else 1
```

```python
step(0.42)
```

```python
intercept_,coef_ = perceptron(X,y)
print(intercept_,coef_)
```
